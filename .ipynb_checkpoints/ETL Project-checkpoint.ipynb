```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
legislators_file = "Resources/legislators-current-terms.csv"
legislators_df = pd.read_csv(legislators_file)
legislators_df.head()
```

```python
no_health_ins_file = "Resources/Population_Without_Health_Insurance_Coverage_State_2013-2015.csv"
no_health_ins_df = pd.read_csv(no_health_ins_file)
no_health_ins_df.head()
```

```python
states_file = "Resources/states.csv"
states_df = pd.read_csv(states_file)
states_df.head()
```

```python
no_health_df = pd.merge(no_health_ins_df, states_df, how = "inner", left_on = "State", right_on = "state_name")
no_health_df.head()
```

```python
new_no_health_df = no_health_df[['abbreviation', '2014 Uninsured - Number']].copy()
new_no_health_df.head()
```

```python
new_no_health_df = new_no_health_df.rename(columns={'abbreviation': 'State', '2014 Uninsured - Number': 'Uninsured Population'})
new_no_health_df.head()
```

```python
legislators_df = legislators_df[['party', 'state']].copy()
legislators_df.head()
```

```python
dem_df = legislators_df.query('party == "Democrat"')
print(len(legislators_df))
print(len(dem_df))
```

```python
dem_df = dem_df.groupby('state').count().reset_index()
dem_df.head()
```

```python
dem_df = dem_df.rename(columns={'state': 'State', 'party': 'Democrats'})
dem_df.head()
```

```python
rep_df = legislators_df.query('party == "Republican"')
print(len(legislators_df))
print(len(rep_df))
```

```python
rep_df = rep_df.groupby('state').count().reset_index()
rep_df.head()
```

```python
rep_df = rep_df.rename(columns={'state': 'State', 'party': 'Republicans'})
rep_df.head()
```

```python
ind_df = legislators_df.query('party == "Independent"')
print(len(legislators_df))
print(len(ind_df))
```

```python
ind_df = ind_df.groupby('state').count().reset_index()
ind_df.head()
```

```python
ind_df = ind_df.rename(columns={'state': 'State', 'party': 'Independent'})
ind_df.head()
```

```python
dem_rep_merge = pd.merge(dem_df, rep_df,  on = "State", how='outer')
dem_rep_merge.head()
```

```python
dem_rep_ind_merge = pd.merge(dem_rep_merge, ind_df,  on = "State", how = "left")
dem_rep_ind_merge.head().sort_values('State')
```

```python
dem_rep_ind_merge = dem_rep_ind_merge.fillna(0)
```

```python
state_df = pd.merge(dem_rep_ind_merge, new_no_health_df,  on = "State", how = "left")

state_df.head().sort_values('State')
```

```python
state_df = state_df.fillna(0)
state_df.head()
```

```python
state_df = state_df.set_index("State")
```

```python
state_df = state_df.drop(['PR', 'GU', 'MP', 'AS', 'VI', 'DC'])
state_df.head()
```

```python
sate_df = state_df.reset_index(inplace=True)
state_df.head()
```

```python
state_df.dtypes
```

```python
state_df['Uninsured Population'] = pd.to_numeric(state_df['Uninsured Population'].str.replace(',', ''))
state_df.head()
```

```python
state_df.dtypes
```

```python
state_df['Uninsured Population'] = state_df['Uninsured Population'] * 1000
state_df.head()
```

```python
state_df['Uninsured Population'] = state_df['Uninsured Population'].astype(int)
state_df['Independent'] = state_df['Independent'].astype(int)
state_df['Republicans'] = state_df['Republicans'].astype(int)
state_df['Democrats'] = state_df['Democrats'].astype(int)
state_df.head()
```
